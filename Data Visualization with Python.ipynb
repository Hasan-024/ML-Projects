Let's discover the use of subplots
To visualize multiple plots together, we can create a figure (overall canvas) and divide it into subplots, each containing a plot. With subplots, we usually work with the artist layer instead of the scripting layer

```python
fig = plt.figure() #Create a figure object named fig

fig1 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
fig2 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot)

#Subplot-1
df_CI.plot(kind='box', color='blue',figsize=(20, 6), ax=fig1) # add to subplot 1
fig1.set_title('Box Plots of Immigrants from China and India (1980 - 2013)')
fig1.set_xlabel('Number of Immigrants')
fig1.set_ylabel('Countries')

#Subplot-2
df_CI.plot(kind='line', figsize=(20, 6), ax=fig2) # add to subplot 2
fig2.set_title ('Line Plots of Immigrants from China and India (1980 - 2013)')
fig2.set_ylabel('Number of Immigrants')
fig2.set_xlabel('Years')
fig2.legend(loc='best')

plt.show()
```

The box plot scans the data and identifies the outliers. In order to be an outlier, the data value must be:

    1. larger than Q3 by at least 1.5 times the interquartile range (IQR), or,
    2. smaller than Q1 by at least 1.5 times the IQR.

Let's look at decade 2000s as an example:

    >Q1 (25%) = 46754.500000
    >Q3 (75%) = 138035.000000
    >IQR = Q3 - Q1 = 91280.5

Using the definition of outlier, any value that is greater than Q3 by 1.5 times IQR will be flagged as outlier.

Outlier > 138035.000000 + (1.5 * 91280.5)
Outlier > 274,955.75

```python
outliers_00s= new_df[new_df['2000s'] > 274955.75]
outliers_00s
```

There are three outlier countries -india, china, philippines as their population for 2000s exceeds the Inter Quartile Range of 274955.75

# Scatter Plot

Using a scatter plot, let's visualize the trend of total immigrantion to Canada (all countries combined) for the years 1980 - 2013.

```python
df_can
```

```python
df_tot = pd.DataFrame(df_can[years])
df_tot
```

```python
df_total = pd.DataFrame(df_can[years].sum(axis=0))
df_total
```

```python
df_total.index = map(int, df_total.index)
```

```python
df_total.reset_index(inplace = True)
```

```python
df_total.columns = ['year', 'total']
```

```python
df_total
```

```python
df_total.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
x = df_total['year']
y = df_total['total']

fit = np.polyfit(x,y, deg=1)
fit
```

```python
df_total.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

# plot line of best fit
plt.plot(x, fit[0] * x + fit[1], color='red') # recall that x is the Years
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()
```
